```python
class NigeriaWeatherFetcher:
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = "http://api.openweathermap.org/data/2.5/weather"
        
        # Major Nigerian cities with coordinates for better accuracy
        self.cities = {
            'Lagos': {'lat': 6.5244, 'lon': 3.3792},
            'Abuja': {'lat': 9.0765, 'lon': 7.3986},
            'Kano': {'lat': 12.0022, 'lon': 8.5920},
            'Ibadan': {'lat': 7.3775, 'lon': 3.9470},
            'Port Harcourt': {'lat': 4.8156, 'lon': 7.0498},
            'Benin City': {'lat': 6.3350, 'lon': 5.6037},
            'Kaduna': {'lat': 10.5222, 'lon': 7.4383},
            'Jos': {'lat': 9.8965, 'lon': 8.8583}
        }
    
    def get_temperature(self, city_name=None, lat=None, lon=None):
        """
        Fetch current temperature for a Nigerian city
        Can use city name from predefined list or custom lat/lon
        """
        params = {
            'appid': self.api_key,
            'units': 'metric'  # Celsius
        }
        
        if city_name and city_name in self.cities:
            params['lat'] = self.cities[city_name]['lat']
            params['lon'] = self.cities[city_name]['lon']
            location = city_name
        elif lat and lon:
            params['lat'] = lat
            params['lon'] = lon
            location = f"Custom ({lat}, {lon})"
        else:
            raise ValueError("Provide either a valid city_name or lat/lon coordinates")
        
        try:
            response = requests.get(self.base_url, params=params)
            response.raise_for_status()
            data = response.json()
            
            return {
                'location': location,
                'temperature': data['main']['temp'],
                'feels_like': data['main']['feels_like'],
                'humidity': data['main']['humidity'],
                'pressure': data['main']['pressure'],
                'description': data['weather'][0]['description'],
                'timestamp': datetime.now().isoformat(),
                'raw_response': data
            }
            
        except requests.exceptions.RequestException as e:
            print(f"Error fetching weather data: {e}")
            return None
        except KeyError as e:
            print(f"Error parsing weather data: {e}")
            return None
    
    def get_all_cities_temperature(self):
        """Fetch temperature for all major Nigerian cities"""
        results = []
        
        for city in self.cities.keys():
            print(f"Fetching weather for {city}...")
            temp_data = self.get_temperature(city)
            if temp_data:
                results.append(temp_data)
            
            # Be nice to the API - small delay between requests
            time.sleep(0.1)
        
        return results
    
    def save_to_csv(self, data, filename=None):
        """Save temperature data to CSV"""
        if not filename:
            filename = f"nigeria_temperature_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        
        # Flatten the data for CSV
        csv_data = []
        for item in data:
            csv_row = {
                'location': item['location'],
                'temperature': item['temperature'],
                'feels_like': item['feels_like'],
                'humidity': item['humidity'],
                'pressure': item['pressure'],
                'description': item['description'],
                'timestamp': item['timestamp']
            }
            csv_data.append(csv_row)
        
        df = pd.DataFrame(csv_data)
        df.to_csv(filename, index=False)
        print(f"Data saved to {filename}")
        return filename
```
